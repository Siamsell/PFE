```
import numpy as np

# Données : superficies (en m²) et prix correspondants (en €)
x = np.array([50, 60, 70, 80])            # superficies
y = np.array([110000, 130000, 150000, 170000])  # prix réels

# Initialisation des paramètres
weight = 1000.0
bias = 2000.0

# Taux d'apprentissage
lr = 0.0001

# Nombre d'itérations
num_iterations = 1000

for i in range(num_iterations):
    # Prédiction : prix prédit = weight * superficie + bias
    predictions = weight * x + bias

    # Calcul de l'erreur
    errors = predictions - y

    # Calcul de la perte (moyenne des erreurs au carré)
    loss = np.mean(errors ** 2)

    # Calcul des gradients
    grad_weight = 2 * np.mean(errors * x)
    grad_bias = 2 * np.mean(errors)

    # Mise à jour des paramètres
    weight = weight - lr * grad_weight
    bias = bias - lr * grad_bias

    if i % 100 == 0:
        print(f"Iteration {i} : perte = {loss:.2f}, poids = {weight:.2f}, biais = {bias:.2f}")

print("\nParamètres finaux :")
print(f"Poids : {weight:.2f}, Biais : {bias:.2f}")
```

```
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Données d'exemple
# x : superficies en m²
# y : prix en milliers d'euros
x = np.array([50, 60, 70, 80], dtype=float)        # superficies
y = np.array([110, 130, 150, 170], dtype=float)     # prix correspondants

# Création du modèle
# Ici, nous construisons un réseau simple avec deux couches cachées
model = Sequential([
    Dense(16, activation='relu', input_shape=[1]),  # Couche cachée avec 16 neurones
    Dense(8, activation='relu'),                      # Couche cachée avec 8 neurones
    Dense(1)                                          # Couche de sortie (prédiction du prix)
])

# Compilation du modèle avec l'optimiseur 'adam' et la fonction de perte MSE
model.compile(optimizer='adam', loss='mean_squared_error')

# Entraînement du modèle
# On entraîne le modèle sur nos données d'exemple pendant 500 itérations (epochs)
model.fit(x, y, epochs=500, verbose=1)

# Utilisation du modèle pour prédire le prix d'une maison
area = 65  # Exemple : une maison de 65 m²
predicted_price = model.predict(np.array([area]))
print(f"Pour une maison de {area} m², le prix prédit est de {predicted_price[0][0]:.2f} milliers d'euros.")
```
